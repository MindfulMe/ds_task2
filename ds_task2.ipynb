Подальший аналіз проводимо лише з відібраними полями.

```python
data_mod = data[["Event Date","Subscription Offer Type","Subscriber ID"]]
```

```python
pd.DataFrame(data_mod)
```

Для подальшого використання дані колонки "Subscription Offer Type" переведено у бінарний формат, де тріал - 0, а підписка - 1.

```python
data_mod['Subscription Offer Type'] = data_mod['Subscription Offer Type'].apply(lambda x: 0 if x == 'Free Trial' else 1)
pd.DataFrame(data_mod)
```

Оскільки ми рахуватимемо лише тих користувачів, які здійснили перехід з тріалу на підписку, то нам необхідно модифікувати наш датафрейм.

```python
data_m0 = data_mod[data_mod['Subscription Offer Type']==0]
data_m1 = data_mod[data_mod['Subscription Offer Type']==1]
data_m = data_m0.merge(data_m1,on='Subscriber ID',how='left')
pd.DataFrame(data_m)
```

На даному етапі краще розібратися з проблемними користувачами, які в той чи інакший період припинили підписку. Таких є 2 (1 користувач тричі відписувався).

```python
pd.DataFrame(data[data["Subscriber ID Reset"]=='Yes'])
```

Цих двох користувачів відразу видалимо з датафрейму. Також видалимо зайві строки, де користувачі не захотіли переходити з тріала на підписку. Таким чином ми отримали датафрейм де присутні лише користувачі, які перейшли на підписку.

```python
#data_m = data_m.loc[data_m['Subscriber ID'] != 223411322456766205]
data_m = data_m.loc[~data_m['Subscriber ID'].isin([223411322456766205,223411322456765550])]
data_m.dropna(subset=['Event Date_y','Subscription Offer Type_y'], inplace=True)
pd.DataFrame(data_m)
```

```python
data_finall = data_m[["Subscriber ID","Event Date_y",]]
pd.DataFrame(data_finall)
```

```python
data_finall = data_finall.groupby('Subscriber ID').count().reset_index() 
data_finall['LTV'] = 6.993 * (1 + data_finall['Event Date_y'])
pd.DataFrame(data_finall)
```
